```python
import numpy as np
import pandas as pd

import torch
torch.manual_seed(1)

X = pd.read_csv('train_X.csv')
y = pd.read_csv('train_Y.csv')
x_test = pd.read_csv('test_X.csv')
submit = pd.read_csv('sample_submit.csv')
```

```python
from imblearn.over_sampling import SMOTE
X_resampled, y_resampled = SMOTE(random_state=1).fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
logist = LogisticRegression()

logist.fit(X_train_std, y_train)
print(logist.score(X_train_std, y_train))
print(logist.score(X_test_std, y_test))
```

```python
sc.fit(X_resampled)
X_std = sc.transform(X_resampled)
x_test_std = sc.transform(x_test)

logist.fit(X_std, y_resampled)

y_pred = logist.predict(x_test_std)
```

```python
submit['label'] = y_pred
submit.to_csv('submit.csv', mode='w', header=True, index=False)
```
