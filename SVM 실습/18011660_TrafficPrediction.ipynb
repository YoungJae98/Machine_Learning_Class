```python

X = pd.DataFrame(pd.read_csv("/kaggle/input/2021-ml-p2/x_train.csv",engine='python',encoding='CP949'))
y = pd.DataFrame(pd.read_csv("/kaggle/input/2021-ml-p2/y_train.csv",engine='python',encoding='CP949'))
x_test = pd.DataFrame(pd.read_csv("/kaggle/input/2021-ml-p2/x_test.csv",engine='python',encoding='CP949'))
submit = pd.read_csv("/kaggle/input/2021-ml-p2/sample_submission.csv")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.svm import SVR
svm = SVR(kernel="linear", C=100)

svm.fit(X_train_std, y_train)

svm.score(X_test_std, y_test)
```

```python
sc.fit(X)
X_std = sc.transform(X)
x_test_std = sc.transform(x_test)

y_pred = svm.predict(x_test_std)

submit['predict'] = y_pred
```

```python
submit.to_csv("submit.csv", index=False)
```
